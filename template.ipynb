## Пример построения графиков в python
Импортируем нужные модули:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Загружаем нужные файлы с помощью `np.loadtxt`:

```python
x, y = np.loadtxt('Практика 01.08\\spectra\\hard\\spectrum1.txt', unpack=True)
x, y # x и y - векторы (массивы) с данными, когда мы пишем так переменную в конце ячейки - выводим ее содержимое
```

```python
data = np.loadtxt('Практика 01.08\\spectra\\hard\\spectrum1.txt')
data # можно загружать многомерные массивы
```

Строим график:

```python
plt.figure(figsize=(16,9)) # открываем новую фигуру заданного размера
plt.plot(x, y) # строим график 
plt.xlabel('Wavelength, $\mu m$', fontsize=22) # Подписываем оси и указываем размер шрифта
plt.ylabel('Transmittance', fontsize=22) # Подписываем оси и указываем размер шрифта
plt.legend(['Пример спектра'], fontsize=22) # Делаем легенду
ax=plt.gca()
ax.tick_params(labelsize=18) # увеличиваем размер чиселок по осям
```

```python
plt.figure(figsize=(16,9)) # открываем новую фигуру заданного размера
plt.plot(data[:,0], data[:,1]) # то же самое, но строим первый и второй стобец 
plt.xlabel('Wavelength, $\mu m$', fontsize=22) # Подписываем оси и указываем размер шрифта
plt.ylabel('Transmittance', fontsize=22) # Подписываем оси и указываем размер шрифта
plt.legend(['Пример спектра'], fontsize=22) # Делаем легенду
ax=plt.gca()
ax.tick_params(labelsize=18) # увеличиваем размер чиселок по осям
```
