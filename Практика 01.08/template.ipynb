## Пример построения графиков в python
Импортируем нужные модули:

```python
import numpy as np
import matplotlib.pyplot as plt
```

Загружаем нужные файлы с помощью `np.loadtxt`:

```python
x, y = np.loadtxt('spectra\spectrum.txt', unpack=True)
x, y # x и y - векторы (массивы) с данными
```

Строим график:

```python
plt.figure(figsize=(16,9)) # открываем новую фигуру заданного размера
plt.plot(x, y) # строим график 
plt.xlabel('Wavelength, $mu m$', fontsize=22) # Подписываем оси и указываем размер шрифта
plt.ylabel('Transmittance', fontsize=22) # Подписываем оси и указываем размер шрифта
plt.legend('Пример ')
```
